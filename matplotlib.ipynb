```python
import matplotlib.pyplot as plt; plt.rcdefaults()
import numpy as np
import matplotlib.pyplot as plt

objects = ('Mexican', 'Pizza', 'Italian', 'Chinese', 'Thai')
y_pos = np.arange(len(objects))
performance = [3.90,3.98,3.88,3.74,4.1]

plt.barh(y_pos, performance, align='center', alpha=0.5)
plt.yticks(y_pos, objects)
plt.xlabel('Rating')
plt.title('Cuisine by Rating')


plt.show()
```

```python
import matplotlib.pyplot as plt
x = [70.2, 84.1, 63.2, 64.5,93.1]
labels = ['Mexican', 'Pizza', 'Italian', 'Chinese', 'Thai']

fig, ax = plt.subplots()
ax.pie(x, labels=labels)
ax.set_title('Not Grubhub Enabled by Cuisine')
plt.tight_layout()

```

```python
fig, ax = plt.subplots(figsize=(6, 6))
ax.pie(x, labels=labels, autopct='%.1f%%')
ax.set_title(' Grubhub Enabled by Cuisine')
plt.tight_layout()

```

```python
fig, ax = plt.subplots(figsize=(6, 6))

x = [70.2, 84.1, 63.2, 64.5,93.1]

ax.pie(x, labels=labels, autopct='%.1f%%',
       wedgeprops={'linewidth': 3.0, 'edgecolor': 'white'},
       textprops={'size': 'x-large'})
ax.set_title(' Grubhub Enabled by Cuisine', fontsize=18)
plt.tight_layout()

```

```python

patches, texts, pcts = ax.pie(
    x, labels=labels, autopct='%.1f%%',
    wedgeprops={'linewidth': 3.0, 'edgecolor': 'white'},
    textprops={'size': 'x-large'})

plt.setp(pcts, color='white', fontweight='bold')
ax.set_title(' Grubhub Enabled by Cuisine', fontsize=18)
plt.tight_layout()
```

```python
fig, ax = plt.subplots(figsize=(6, 6))

patches, texts, pcts = ax.pie(
    x, labels=labels, autopct='%.1f%%',
    wedgeprops={'linewidth': 3.0, 'edgecolor': 'white'},
    textprops={'size': 'x-large'},
    startangle=90)

for i, patch in enumerate(patches):
  texts[i].set_color(patch.get_facecolor())
plt.setp(pcts, color='white')
plt.setp(texts, fontweight=600)
ax.set_title('Grubhub Enabled by Cuisine', fontsize=18)
plt.tight_layout()

```
